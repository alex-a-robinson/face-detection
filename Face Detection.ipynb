```python
import urllib
import os

def download(url, name, extension):
    '''Downloads and saves an image if it dosen't already exsist
    '''
    if not os.path.isfile('data/%s.%s' % (name, extension)):
        with open('data/%s.%s' % (name, extension), 'wb') as image_file:
            image_file.write(urllib.urlopen(url).read())
            image_file.close()
```

```python
face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('data/haarcascade_eye.xml')
mouth_cascade = cv2.CascadeClassifier('data/haarcascade_mcs_mouth.xml')

def detect_face(img, minNeighbours=10, 
                eye_detection=True,
                mouth_detection=True):
    '''Detects a face, eyes and mouth in an image returning the image with features outlined
    '''
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, minNeighbors=minNeighbours)

    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 4)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]

        if not eye_detection:    
            break
        
        # Detect eyes
        eyes = eye_cascade.detectMultiScale(roi_gray, minNeighbors=minNeighbours)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 4)

        if not mouth_detection:    
            break
        # Detect mouth(s)
        mouths = mouth_cascade.detectMultiScale(roi_gray, minNeighbors=minNeighbours)
        for (mx, my, mw, mh) in mouths:
            cv2.rectangle(roi_color, (mx, my), (mx+mw, my+mh), (0, 0, 255), 4)
            
    return img

def detect_face_helper(path):
    '''Helper function for detect face
    '''
    img = cv2.imread(path)
    face = detect_face(img.copy(), 5)
    
    plt.figure()
    plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
```
